Lineal 
$f_1(x) =  ax$

Cuadrática

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Ejemplos Ilustrativos

## Ejemplo 1:  Encontrar parámetros de la función lineal $f_1$

 Construcción del problema

```python
# definimos nuestro dominio
x = np.arange(-5,5+1)

#definimos el contradominio con un parámetro $a=4$ 
a = 4
y = a*x

#Agregamos ruido uniformemente distribuido 

y_noise = y +np.random.rand(len(x))*20-10

plt.plot(y)
plt.plot(y_noise,'ro')
```

Definimos la función de costo cuadrática que queremos minimizar.

$C(a) = \sum_{i=1}^m 0.5(y_i-a \cdot x_i)^2 $

Y necesitamos su derivada

$\frac{\delta C_i }{\delta a} = -(y_i-ax_i) x_i$

```python
#La programación de la derivada de la función de costo 

def pd(x,y,a):
    s = 0
    for i in np.arange(len(x)):
        dif = -(y[i]-a*x[i])
        s = s + dif*(x[i])
    return s
```

### Gradiente Descendente

```python
#condicion inicial del parámetro
av = 10
lav = list([av])
#tamaño de paso
n = 0.001

for it in np.arange(10000):
    
    
    #derivada parcial
    dp = pd(x,y_noise,av)
    
    an = av-n*dp
    av = an
    lav.append(av)

print("Estimación")
print("a =",lav[-1])
print("Real")
print("a =",a)
```

# Gráfica de Convergencia 

```python
plt.plot(lav[:50])
```

```python
plt.plot(lav[-1]*x)
plt.plot(a*x)
plt.plot(y_noise)
plt.show()
```

## Ejemplo 1:  Encontrar parámetros de la función cuadrática $f_2$
$f_2(x) = 10x^2$

```python

# función cuadrática con parámetro a=10
a = 10
y = a*x**2

#Agregamos ruido uniformemente distribuido 

y_noise = y +np.random.rand(len(x))*100-50

plt.plot(y)
plt.plot(y_noise,'ro')
plt.show()
```

Definimos la función de costo cuadrática.

$C(a) = \sum_{i=1}^m 0.5(y_i-a \cdot x_i^2)^2 $

Y necesitamos su derivada

$\frac{\delta C_i }{\delta a} = -(y_i-ax_i^2) x_i^2$

```python
#Generamos la función de evaluación usando la derivada de la función de costo

def pd(x,y,a):
    s = 0
    for i in np.arange(len(x)):
        dif = (y[i]-a*x[i]**2)
        s = s + dif*((x[i])**2)
    return s
```

```python
#condicion inicial del parámetro, propuesta inicial
av = 30
lav = list([av])

for it in np.arange(1000):
    #tamaño de paso
    n = 0.0001
    
    #derivada parcial
    dp = pd(x,y_noise,av)
    
    an = av+n*dp
    av = an
    lav.append(av)

print("Estimación")
print("a =",lav[-1])
print("Real")
print("a =",a)
```

```python
plt.plot(lav[:100])
plt.show()
```

```python
#comparación curva estimada con la real
plt.plot(lav[-1]*x**2)
plt.plot(a*x**2)
plt.plot(y_noise,'ro')
plt.show()
```
